```python
# ============================================
# 1. 입력 파일 경로 설정 ★★★
# ============================================
input_path = r"C:\Users\itwill\Desktop\불용어사전\올리브영_가나다순_저장완료.txt" 

# ============================================
# 2. 불용어 파일 경로 설정 ★★★ (이거 추가됨!)
# ============================================
stopword_path = r"C:\Users\itwill\Desktop\불용어사전\stopword 최종.txt" 
# ↑ 여기에 불용어 리스트 txt 파일 경로 넣기


# ============================================
# 3. 원본 텍스트 읽기
# ============================================
with open(input_path, "r", encoding="utf-8") as f:
    all_keywords_text = f.read()   # 전체 텍스트 읽기


# ============================================
# 4. 불용어 파일 읽기
# ============================================
with open(stopword_path, "r", encoding="utf-8") as f:
    stopwords = set(line.strip() for line in f if line.strip())  
    # ★ 한 줄씩 읽어서 공백 제거 → set()에 저장(검색 빠르게)


# ============================================
# 5. 공백 기준으로 단어 분리
# ============================================
keywords_list = all_keywords_text.split()


# ============================================
# 6. 불용어 제거 (여기서 stopword 반영)
# ============================================
filtered_keywords = [kw for kw in keywords_list if kw not in stopwords]


# ============================================
# 7. 중복 제거 + 가나다순 정렬
# ============================================
unique_sorted_keywords = sorted(set(filtered_keywords))


# ============================================
# 8. 출력 파일 경로 ★★★
# ============================================
output_path = r"C:\Users\itwill\Desktop\중간프로젝트 코스메틱\clean_sorted_5.txt"


# ============================================
# 9. 결과 저장 (단어 1개당 1줄)
# ============================================
with open(output_path, "w", encoding="utf-8") as f:
    for kw in unique_sorted_keywords:
        f.write(kw + "\n")


print("완료! 정리된 파일 저장됨 →", output_path)
print("총 단어 수:", len(unique_sorted_keywords))
```
