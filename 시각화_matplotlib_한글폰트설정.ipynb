```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
import pandas as pd

#파일 불러오기 
data = pd.read_csv(r"C:\py_temp\과일3개.csv", encoding = 'cp949')

#데이터 구조확인 
print(data.head())
print(data.columns)
```

```python
#한글폰트설정(그래프에 한글이 깨지지않도록 폰트설정)
font_location = r"C:\Windows\Fonts\batang.ttc" 
font_name = fm.FontProperties(fname =font_location).get_name()
matplotlib.rc('font', family=font_name) #전부 폰트적용해라

#마이너스 기호 깨짐방지코드 써주기
plt.rcParams['axes.unicode_minus']=False

#그래프 스타일 및 크기설정 
plt.style.use('ggplot')
plt.figure( figsize = (8,5) ) #1번째로 도화지꺼냄

#데이터를 추출한다
data_1 = data[data['종류']=='수박']['무게_g']
data_2 = data[data['종류']=='자두']['무게_g']
data_3 =data[data['종류']=='참외']['무게_g']

#라인그래프로 표현하기
plt.plot(
    data_1.reset_index(drop=True),
    label='수박',
    marker ='o'
)
plt.plot(
    data_2.reset_index(drop=True),
    label='자두',
    marker = 's'
)
plt.plot(
    data_3.reset_index(drop=True),
    label='참외',
    marker='^'
)
# 표 제목,축라벨,범례넣기
plt.title('3가지 과일 종류별 무게 비교') #그래프전체제목
plt.xlabel('번호')   #x축제목
plt.ylabel('무게')   #y축제목

plt.legend(loc='best')
plt.show()






    
# plt.xlim(0,4)안써도되는게 matplotlib이 csv파일을읽고 자동을로 축범위를 계산해서 설정한다
#plt.ylim(500,2500)
#plt.legend()
```
